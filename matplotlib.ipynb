```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
tips=sns.load_dataset('tips')
```

```python
tips.head()
```

```python
tips.shape
```

```python
plt.plot()
```

```python
plt.plot([1,2,3,4,5,6],[1,5,8,3,6,25])
```

```python
plt.title('tips graphs')
plt.xlabel('X-Axis')
plt.ylabel('Y-Axis')
```

```python
# DISPLOT
sns.distplot(tips['tip'])
```

```python
sns.displot(tips['total_bill'],kde=False,bins=30)
```

```python
sns.displot(tips['total_bill'],kde=True,bins=25)
```

```python
sns.displot(tips['total_bill'],kde=True,bins=25,color='g')
```

```python
#jointplot digram commands 
# *scatter,*resid,*hex,*reg,*kde


sns.jointplot(x=tips['total_bill'],y=tips['tip'],kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='reg')
```

```python
sns.jointplot(x='total_bill',y='tip',data=tips,kind='scatter')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='resid')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='hex')
sns.jointplot(x='total_bill',y='tip',data=tips,kind='kde')
```

```python
#regplot
sns.regplot(x='total_bill',y='tip',data=tips)
```

```python
#pairplot
sns.pairplot(tips)
```

```python
sns.pairplot(tips,hue='sex')
```

```python
#boxplot
sns.boxplot(tips['total_bill'])
```

```python
sns.boxplot(tips['sex'],tips['total_bill'])
```

```python
sns.boxplot(tips['size'],tips['total_bill'])
```

```python
sns.boxplot(tips['smoker'],tips['total_bill'],hue=tips['sex'])
```

```python
#Countplot

sns.countplot('time',data=tips,hue='size')
```

```python
sns.countplot('time',data=tips)
```

```python
sns.countplot('time',data=tips,hue='sex')
```
